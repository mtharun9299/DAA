```python
import random
import sys
import time # Added import for the 'time' module

def interpolation_search(arr, target):
    """
    Interpolation Search Algorithm
    Time Complexity: O(log log n) average, O(n) worst case
    Space Complexity: O(1)
    """
    low, high = 0, len(arr) - 1
    comparisons = 0
    while low <= high and arr[low] <= target <= arr[high]:
        comparisons += 1
        if low == high:
            if arr[low] == target:
                return low, comparisons
            # Corrected indentation for this return statement
            return -1, comparisons

        # Interpolation formula
        if arr[high] == arr[low]: # Avoid division by zero if range is flat
            if arr[low] == target:
                return low, comparisons
            else:
                return -1, comparisons
        pos = low + int(((target - arr[low]) * (high - low)) /
                        (arr[high] - arr[low]))

        # Ensure pos is within bounds
        if not (low <= pos <= high):
            # If pos is out of bounds, target is not in the remaining range
            return -1, comparisons

        if arr[pos] == target:
            return pos, comparisons
        elif arr[pos] < target:
            low = pos + 1
        else:
            high = pos - 1
    return -1, comparisons

def binary_search(arr, target):
    """Binary Search for comparison"""
    low, high = 0, len(arr) - 1
    comparisons = 0
    while low <= high:
        comparisons += 1
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid, comparisons
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    # Corrected indentation for this return statement
    return -1, comparisons

def performance_analysis():
    sizes = [1000, 5000, 10000, 50000, 100000]
    print(f"{'Size':>10} {'IS Time(ms)':>14} {'BS Time(ms)':>14} "
          f"{'IS Comparisons':>16} {'BS Comparisons':>16}")
    print('-' * 75)
    for size in sizes:
        arr = sorted(random.sample(range(size * 10), size))
        target = arr[random.randint(0, size - 1)]

        # Interpolation Search timing and comparison accumulation
        start_is = time.perf_counter()
        comp_is_total = 0
        for _ in range(100):
            idx_is, comp_is = interpolation_search(arr, target)
            comp_is_total += comp_is
        is_time = (time.perf_counter() - start_is) / 100 * 1000
        comp_is_avg = comp_is_total / 100

        # Binary Search timing and comparison accumulation
        start_bs = time.perf_counter()
        comp_bs_total = 0
        for _ in range(100):
            idx_bs, comp_bs = binary_search(arr, target)
            comp_bs_total += comp_bs
        bs_time = (time.perf_counter() - start_bs) / 100 * 1000
        comp_bs_avg = comp_bs_total / 100

        print(f"{size:>10} {is_time:>14.4f} {bs_time:>14.4f} "
              f"{comp_is_avg:>16.2f} {comp_bs_avg:>16.2f}")

# --- Main ---
arr = [2, 5, 10, 15, 23, 35, 48, 60, 75, 90, 105, 120]
target = 35
idx, comps = interpolation_search(arr, target)
print(f"Array: {arr}")
print(f"Searching for: {target}")
print(f"Found at index: {idx}, Comparisons: {comps}")
print()
performance_analysis()
```
